```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import random
import os
```

```python
# import main dataset
csvfile = "data/GB_youtube_trending_data.csv"
df = pd.read_csv(csvfile)

df.head()
```

```python
# filter data by 2020 date
df[(df['publishedAt'] > '2020-09-01') & (df['publishedAt'] < '2020-09-30')]
```

```python
# create new dataset from filter
df1 = df[(df['publishedAt'] > '2020-09-01') & (df['publishedAt'] < '2020-09-30')]
df1r = df1.drop_duplicates(subset=['video_id'])
```

```python
# save dataset
df1.to_csv('new data/2020_data2.csv', index=False)
df1r.to_csv('new data/2020_data_id.csv', index=False)
```

```python
# repeat for 2022
df[(df['publishedAt'] > '2022-09-01') & (df['publishedAt'] < '2022-09-30')]
df2 = df[(df['publishedAt'] > '2022-09-01') & (df['publishedAt'] < '2022-09-30')]
df2r = df2.drop_duplicates(subset=['video_id'])
df2.to_csv('new data/2022_data.csv', index=False)
df2r.to_csv('new data/2022_data_id.csv', index=False)
```
